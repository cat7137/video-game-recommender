```python
video_games = vgs[['Unnamed: 0', 'name', 'plot', 'Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy', 'Mystery', 'Sci-Fi', 'Thriller']]
video_games
```

```python
def get_genres(row):
    genres = []
    if row['Action']:
        genres.append('Action')
    if row['Adventure']:
        genres.append('Adventure')
    if row['Comedy']:
        genres.append('Comedy')
    if row['Crime']:
        genres.append('Crime')
    if row['Family']:
        genres.append('Family')
    if row['Fantasy']:
        genres.append('Fantasy')
    if row['Mystery']:
        genres.append('Mystery')
    if row['Sci-Fi']:
        genres.append('Sci-Fi')
    if row['Thriller']:
        genres.append('Thriller')
    return genres

video_games['genres'] = video_games.apply(get_genres, axis=1)
    
```

```python
video_games = video_games.drop(columns=['Action', 'Adventure', 'Comedy', 'Crime', 'Family','Fantasy', 'Mystery', 'Sci-Fi', 'Thriller'])
```

```python
video_games
```

```python
video_games['plot'] = video_games['plot'].apply(lambda x:x.split())
```

```python
video_games.head(2)
```

```python
video_games['tags'] = video_games['plot'] + video_games['genres']
video_games.head()
```

```python
new_df = video_games[['name','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['name'] == "Spider-Man"].index[0]
```

```python
def recommend(video_game):
    index = new_df[new_df['name'] == video_game].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x:x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]]['name'])
```

```python
recommend('Batman')
```
